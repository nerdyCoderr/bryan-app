```python
import threading
import cv2
from collections import deque

def image_generator(d):
  # Open the camera
  cap = cv2.VideoCapture(0)
  
  # Run in a loop indefinitely
  while True:
    # Read an image from the camera
    ret, image = cap.read()
    
    # Add the image to the end of the deque
    d.append(image)

# Create a deque
d = deque(maxlen=3)

# Create a thread and start it
thread = threading.Thread(target=image_generator, args=(d,))
thread.start()

# Wait for keyboard input in the main thread
while True:
  user_input = input("Enter something: ")
  
  # If the user presses 's', get the image from the deque
  if user_input == 's':
    # Remove and retrieve the image from the end of the deque
    image = d.pop()
    
    # Display the image using cv2
    cv2.imshow("Image", image)
    cv2.waitKey(0)
  else:
    print("You entered:", user_input)
```
